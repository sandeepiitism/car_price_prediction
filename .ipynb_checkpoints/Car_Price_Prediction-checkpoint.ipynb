## EDA

### Cleaning year column and changing to int type

```python
df=df[df['year'].str.isnumeric()]
```

```python
df.dtypes
```

```python
df['year']=df['year'].astype(int)
df.dtypes
```

### Cleaning price column and changing to int type

```python
df=df[df['Price']!='Ask For Price']
```

```python
df['Price']=df['Price'].str.replace(',','').astype(int)
```

```python
df['kms_driven']=df['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
df=df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
df.dtypes
```

### Cleaning fuel_type column and changing to int type

```python
df=df[~df['fuel_type'].isna()]
```

```python
df.shape
```

### Cleaning car name column

```python
df['name']=df['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
df=df.reset_index(drop=True)
```

### Model Development

```python
X=df[['name','company','year','kms_driven','fuel_type']]
y=df['Price']
```

```python
X
```

```python
y.shape
```

### Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

#### Creating an OneHotEncoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
### column_transformer for categorical features
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

#### Linear Regression Model

```python
lr=LinearRegression()
```

#### Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

#### Fitting the  model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

#### Checking R2 Score

```python
r2_score(y_test,y_pred)
```

#### Train_test Split with random Sample

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

#### The best model is found at a certain random state 

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
